```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/titanicdataset.csv')
```

```python
df.head()
```

Distribution plots


a) Dist plot

```python
sns.distplot(x = df['Age'], bins = 10)
```

```python
sns.distplot(df['Age'], bins = 10,kde=False)
```

b) Joint plot

```python
sns.jointplot(x = df['Age'], y = df['Fare'], kind =
'scatter')
```

```python
sns.jointplot(x = df['Age'], y = df['Fare'], kind = 'hex')
```

c) Rug plot

```python
sns.rugplot(df['Fare'])
```

Categorical plots


a) Bar plot

```python
sns.barplot(x='Sex', y='Age', data=df)
```

b) Count plot

```python
sns.countplot(x='Sex', data=df)
```

c) Box plot

```python
sns.boxplot(x='Sex', y='Age', data=df)
```

```python
sns.boxplot(x='Sex', y='Age', data=df, hue="Survived")
```

d) Violin plot

```python
sns.violinplot(x='Sex', y='Age', data=df)
```

```python
sns.violinplot(x='Sex', y='Age', data=df, hue='Survived')
```

Advanced plots

a) Strip plot

```python
sns.stripplot(x='Sex', y='Age', data=df, jitter=False)
```

```python
sns.stripplot(x='Sex', y='Age', data=df, jitter=True)
```

```python
sns.stripplot(x='Sex', y='Age', data=df, jitter=True, hue='Survived')
```

b) Swarm plot

```python
sns.swarmplot(x='Sex', y='Age', data=df)
```

```python
sns.swarmplot(x='Sex', y='Age', data=df, hue='Survived')
```

Matrix plots


Heat map

```python
numerical_df = df.select_dtypes(include=np.number)
corr = numerical_df.corr()
sns.heatmap(corr)
plt.show()
```

Checking how the price of the ticket (column name: 'fare') for each passenger is
