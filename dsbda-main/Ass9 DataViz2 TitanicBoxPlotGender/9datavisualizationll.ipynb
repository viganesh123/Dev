```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

Count plot

```python
print(df['Survived'].value_counts())
# plt.show()
```

Pie chart

```python
df['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

Histogram

```python
plt.hist(df['Age'], bins=5)
plt.show()
```

Distplot

```python
sns.distplot(df['Age'],kde=True)
plt.show()
```

Scatter plot

Bar plot

```python
sns.barplot(x=df['Pclass'], y=df['Age'])
plt.show()
```

```python
sns.barplot(x=df['Pclass'], y=df['Fare'], hue=df["Sex"])
plt.show()
```

```python
sns.boxplot(x=df['Sex'], y=df["Age"], hue=df["Survived"])
plt.show()
```

Distplot

```python
sns.distplot(df[df['Survived'] == 0]['Age'], hist=False, color="blue")
sns.distplot(df[df['Survived'] == 1]['Age'], hist=False, color="orange")
plt.show()
```

Heat map

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'], df['Survived']))
```

```python
sns.clustermap(pd.crosstab(df['Parch'], df['Survived']))
plt.show()
```

```python
plt.figure(24,6)
sns.countplot(x='Age', data=df, hue='Survived')
```
